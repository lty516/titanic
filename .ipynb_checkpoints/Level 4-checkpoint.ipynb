```python
import pandas as pd
```

# Load Dataset

```python
train = pd.read_csv("data/titanic/train.csv")
```

```python
train.head()
```

```python
test= pd.read_csv('data/titanic/test.csv')
```

```python
test.head()
```

## Explore

```python
import seaborn as sns
```

```python
sns.countplot(data=train, x='Embarked',hue="Survived")
```

```python
sns.countplot(data= train, x='Pclass',hue='Survived')
```

```python
low_fare = train[train["Fare"] < 100]
sns.lmplot(data=low_fare, x='Age',y='Fare',hue='Survived', fit_reg=False)
```

## Preprocessing


#### Encode Sex

```python
train["Sex_encode"]= train["Sex"].replace("male",0).replace("female",1)

train[['Sex','Sex_encode']].head()
```

```python
test["Sex_encode"]= test["Sex"].replace("male",0).replace("female",1)

test[['Sex','Sex_encode']].head()
```

### Fill in missing Fare

```python
test["Fare"] = test["Fare"].fillna(0)
# train Fare nan 
# train[train["Fare"].isnull()]
test[test["Fare"].isnull()]
```

##  Train

```python
# Feature(x) - Pclass, Sex, Fare, ...
# Label(y)  - Survived
```

```python
feature_names =['Pclass',"Sex_encode","Fare"]
feature_names
```

```python
X = train[feature_names]
X.head()
```

```python
label_name = "Survived"

y = train[label_name]
y.head()
```

## Use Decision Tree

```python
#scikit-learn == sklearn
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 5)
model
```

```python
# 1) fit(train) ->  2) predict(test)
model.fit(X,y)
```

# Visualize

```python
import graphviz
from sklearn.tree import export_graphviz

tree = export_graphviz(model,
                      feature_names = feature_names,
                      class_names = ["perish","Survived"])

graphviz.Source(tree)
```

### Predict

```python
# test
X_test = test[feature_names]
X_test.head()
```

```python
prediction_list = model.predict(X_test)
prediction_list[0:5]
```

### Submit

```python
submit = pd.read_csv('data/titanic/gender_submission.csv')

submit["Survived"] = prediction_list

submit.head()
```

```python
submit.to_csv('decision-tree.csv',index = False)
```
