```python
import pandas as pd
```

```python
file_url = "https://bit.ly/ds01-weight-history"

data = pd.read_csv(file_url,index_col ="회차")
data
```

```python
data[["지점"]].describe()
```

```python
data["지점"]= data["지점"].replace("강남구","강남")
data
```

```python
#NaN, null
data["몸무게"] = data["몸무게"].fillna(75.8)
data
```

```python
data['몸무게(prev)'] = data["몸무게"].shift(+1)
data['몸무게(next)'] = data["몸무게"].shift(-1)
data
```

```python
data["측정일"] = pd.to_datetime(data["측정일"])
data

data['측정일'].dt.year
```

```python
data['측정일'].dt.month
```

## 컴첨 추가 & 수정

```python
data["지역"]= "부산"
data
```

```python
data["결제"] = [True,True,True,True,False,True,False,True,True]
data
```

```python
data["몸무게(gram)"] = data["몸무게"] *1000
data
```

```python
data["과체중"] = data["몸무게"] > 75.8
data
```

```python
data.loc[data["몸무게"]>75.8,"상탱"] = "과체중"
data.loc[data["몸무게"]<=75.8,"상탱"] = "정상"

data
```

## 컬럼 삭제

```python
data=  data.drop("몸무게(prev)", axis ="columns")
data
```

```python
data = data.drop(["몸무게(next)","몸무게(gram)"],axis= "columns")
data
```
